# Analysis Code for MTG Price Watcher

```python
#imports 
import pandas as pd
import numpy as np
import price_analysis
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = price_analysis.read_price_data()
```

```python
df = price_analysis.add_moving_avg(df, return_last=False)
df["checked_timestamp"] = pd.to_datetime(df["checked_timestamp"])
```

```python
f, ax = plt.subplots()
f.set(figheight=8, figwidth=8)
ax.set(title="Price over time",
        xlabel="Date",
        ylabel="Price (USD)")
sns.lineplot(x="checked_timestamp",
             y="price_usd",
             hue="card_name",
             data=df)
plt.xticks(rotation=45)
plt.show()
```
